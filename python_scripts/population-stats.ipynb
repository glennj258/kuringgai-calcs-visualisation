```python
# migration between
def get_migration(from_location, to_location, age_range):
    to_key = f"{to_location}***{age_range}"

    return internal_mig_t[from_location][to_key]

# internal_mig_t["Greater Sydney"]["Greater Melbourne***20-29 years"]#["GCCSA***AGE10P Age in Ten Year Groups"]# == "Greater Melbourne***20-29 years"]
```

```python
def get_net_migration(from_location, to_location, age_range):
    return get_migration(from_location, to_location, age_range) - get_migration(to_location, from_location, age_range)
```

```python
age_ranges = ["0-9 years", "10-19 years", "20-29 years", "30-39 years", "40-49 years", "50-59 years", "60-69 years","70-79 years", "80-89 years", "90-99 years"]
```

```python
def get_age_mig_distribution(from_location, to_location):
    # print(f"{from_location} to {to_location}")
    total_migation = 0
    for age_range in age_ranges:
        count_net_migation = get_net_migration(from_location, to_location, age_range)
        print(age_range, count_net_migation)
        total_migation += count_net_migation
    
    print("Total: ", total_migation)
```

```python
net_mig = {}

for from_gcc in GCCSAs:
    net_mig[f"{from_gcc}"] = {}
    count_total_gcc = 0
    for to_gcc in GCCSAs:
        net_mig[f"{from_gcc}"][f"{to_gcc}"] = {}
        count_total_ages = 0
        for age_range in age_ranges:
            net_mig[f"{from_gcc}"][f"{to_gcc}"][f"{age_range}"] = str(get_net_migration(to_gcc, from_gcc, age_range))
            count_total_ages += get_net_migration(to_gcc, from_gcc, age_range)

        count_total_gcc += count_total_ages
        net_mig[f"{from_gcc}"][f"{to_gcc}"]["Total"] = str(count_total_ages)
    net_mig[f"{from_gcc}"]["Total"] = str(count_total_gcc)
```
